## Sample Data

### Estimation

```python
from sklearn.datasets import make_regression
from sklearn.metrics import mean_squared_error
```

```python
X, y = make_regression(n_samples=2500, n_features=5, n_informative=5,
                       random_state=500, noise=10)
```

```python
X[:7]
```

```python
y[:7]
```

```python
reg = np.linalg.lstsq(X, y, rcond=-1)[0]
reg
```

```python
p = np.dot(X, reg)
```

```python
mean_squared_error(y, p)
```

```python
from sklearn.neural_network import MLPRegressor
```

```python
model = MLPRegressor(hidden_layer_sizes=[128, 128, 128],
                    max_iter=5000)
```

```python
%time model.fit(X, y)
```

```python
p = model.predict(X)
```

```python
mean_squared_error(y, p)
```

```python
X[:, 3].mean()
```

```python
X[:, 3].std()
```

### Classification

```python
from sklearn.datasets import make_classification
from sklearn.metrics import accuracy_score
```

```python
X, y = make_classification(n_samples=250, n_features=2, n_redundant=0,
                    random_state=100)
```

```python
X[:7]
```

```python
y[:7]
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm');
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model = GaussianNB()
```

```python
model.fit(X, y)
```

```python
p = model.predict(X)
```

```python
accuracy_score(y, p)
```

```python
Xc = X[y == p]
Xf = X[y != p]
```

```python
plt.scatter(Xc[:, 0], Xc[:, 1], c=y[y == p],
            cmap='coolwarm', marker='o')
plt.scatter(Xf[:, 0], Xf[:, 1], c=y[y != p],
            cmap='coolwarm', marker='x');
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
model_dnn = MLPClassifier(hidden_layer_sizes=[128, 128],
                      max_iter=5000)
```

```python
%time model_dnn.fit(X, y)
```

```python
p = model_dnn.predict(X)
```

```python
accuracy_score(y, p)
```

```python
Xc = X[y == p]
Xf = X[y != p]
```

```python
plt.scatter(Xc[:, 0], Xc[:, 1], c=y[y == p],
            cmap='coolwarm', marker='o')
plt.scatter(Xf[:, 0], Xf[:, 1], c=y[y != p],
            cmap='coolwarm', marker='x');
```

```python
X, y = make_classification(n_samples=1000, n_features=100,
                           n_informative=10, n_redundant=0,
                           random_state=100)
```

```python
model.fit(X, y)
```

```python
p = model.predict(X)
```

```python
accuracy_score(y, p)
```

```python
%time model_dnn.fit(X, y)
```

```python
p = model_dnn.predict(X)
```

```python
accuracy_score(y, p)
```
