#### Functional Relationship

```python
x = np.linspace(-5, 5)
```

```python
y = x ** 2 / 2 + np.sin(x) * 2
```

```python
plt.plot(x, y);
```

```python
reg = np.polyfit(x, y, deg=12)
```

```python
reg
```

```python
p = np.polyval(reg, x)
p[:7]
```

```python
plt.plot(x, y)
plt.plot(x, p, 'ro');
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y, p)
```

```python
from sklearn.neural_network import MLPRegressor
```

```python
model = MLPRegressor(hidden_layer_sizes=[100, 100, 100, 100],
                     learning_rate_init=0.005,
                     max_iter=2000)
```

```python
# MLPRegressor?
```

```python
%time model.fit(x.reshape(-1, 1), y)
```

```python
p = model.predict(x.reshape(-1, 1))
```

```python
plt.plot(x, y)
plt.plot(x, p, 'ro');
```

```python
mean_squared_error(y, p)
```

### Random Data

```python
x = np.linspace(-5, 5, 15)
```

```python
from numpy.random import default_rng
```

```python
rng = default_rng()
```

```python
y = rng.standard_normal(len(x)) * 2
```

```python
plt.plot(x, y);
```

```python
reg = np.polyfit(x, y, deg=14)
```

```python
reg
```

```python
p = np.polyval(reg, x)
```

```python
plt.plot(x, y)
plt.plot(x, p, 'ro');
```

```python
mean_squared_error(y, p)
```

```python
model = MLPRegressor(hidden_layer_sizes=[100, 100, 100, 100],
                     learning_rate_init=0.00025,
                     max_iter=3000)
```

```python
%time model.fit(x.reshape(-1, 1), y)
```

```python
p = model.predict(x.reshape(-1, 1))
```

```python
plt.plot(x, y)
plt.plot(x, p, 'ro');
```

```python
mean_squared_error(y, p)
```
