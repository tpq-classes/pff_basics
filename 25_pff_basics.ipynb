## Sample Data

```python
from sklearn.datasets import make_regression
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
# make_regression?
```

```python
X, y = make_regression(n_samples=250, n_features=5,
                       n_informative=5, n_targets=1,
                       random_state=500, noise=10)
```

```python
X[:7]
```

```python
y[:7]
```

### OLS Regression

```python
reg = np.linalg.lstsq(X, y, rcond=-1)[0]
```

```python
reg
```

```python
np.dot(X, reg)[:7]
```

```python
p = np.dot(X, reg)
```

```python
mean_squared_error(y, p)
```

```python
mean_absolute_error(y, p)
```

```python
plt.plot(y)
plt.plot(p, alpha=0.6);
```

```python
plt.bar(np.arange(len(y)), p - y);
```

```python
plt.hist(p - y, bins=25);
```

### Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()  # 1. step
```

```python
model.fit(X, y)  # 2. step
```

```python
p = model.predict(X)  # 3. step
p[:7]
```

```python
mean_squared_error(y, p)
```

```python
mean_absolute_error(y, p)
```

```python
plt.plot(y)
plt.plot(p, alpha=0.6);
```

```python
plt.bar(np.arange(len(y)), p - y);
```

```python
plt.hist(p - y, bins=25);
```
