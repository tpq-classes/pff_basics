```python
N = 1000000
```

```python
r = range(N)
```

```python
%time l = list(range(N))
```

```python
sys.getsizeof(l)
```

```python
%time t = tuple(range(N))
```

```python
sys.getsizeof(t)
```

```python
%%time
s = 0
for i in range(N):
    s += i
s
```

```python
%%time
s = 0
for i in l:
    s += i
s
```

```python
%%time
s = 0
for i in t:
    s += i
s
```

```python
def isum(o):
    s = 0
    for i in o:
        s += i
    return s
```

```python
%timeit isum(r)
```

```python
%timeit isum(l)
```

```python
%timeit isum(t)
```

```python
%timeit sum(r)
```

```python
%timeit sum(l)
```

```python
%timeit sum(t)
```
