## Arbitrage Pricing

## Financial Market

```python
B0 = 10
```

```python
S0 = 10
```

```python
B = np.array((11, 11))
```

```python
B
```

```python
S = np.array((20, 5))
```

```python
S
```

```python
M0 = np.array((B0, S0))
```

```python
M0
```

```python
M = np.array((B, S)).T
```

```python
M
```

## Returns

```python
R = M / M0 - 1
```

```python
R
```

## Expected Return

```python
P = np.array((0.5, 0.5))
```

```python
P.sum()
```

```python
# np.dot?
```

```python
np.dot(P, S)  # expected payoff
```

```python
np.dot(P, B)  # expected payoff
```

```python
np.dot(P, S) / S0  # gross expected return
```

```python
np.dot(P, S) / S0 - 1  # net expected return
```

```python
np.dot(P, B) / B0  # gross expected return
```

```python
np.dot(P, B) / B0 - 1  # net expected return
```

```python
np.dot(P, R)  # net expected return
```

```python
R.mean(axis=0)  # assumes equal probability
```

## Risk/Volatility

```python
R.std(axis=0)  # assumes equal probability
```

## Contingent Claim

```python
K = 14.5
```

```python
C = np.maximum(S - K, 0)
```

```python
C
```

```python
# C0 = ?
```

## Pricing via Discounting 

**Wrong Way**.

```python
i = 0.1  # risk-less rate (bond)
```

```python
C0 = 1 / (1 + i) * np.dot(P, C)
```

```python
C0
```

```python
i = 0.25  # risky rate (stock)
```

```python
C0 = 1 / (1 + i) * np.dot(P, C)
```

```python
C0
```

## Pricing by Arbitrage

```python
C
```

```python
M
```

```python
# np.linalg.solve?
```

```python
phi = np.linalg.solve(M, C)  # replication of the payoff
```

```python
phi
```

```python
np.dot(phi, M.T)
```

```python
C0 = np.dot(phi, M0)  # arbitrage-free price/value
```

```python
C0
```

## Pricing via Discounting 

**Correct Way**.

```python
i = (C / C0 - 1).mean()
```

```python
i
```

```python
C0 = 1 / (1 + i) * np.dot(P, C)
```

```python
C0
```

## Pricing via OLS Regression

```python
# np.linalg.lstsq?
```

```python
phi = np.linalg.lstsq(M, C, rcond=-1)[0]
```

```python
phi
```

```python
np.dot(phi, M.T)
```

```python
np.dot(M, phi)
```
