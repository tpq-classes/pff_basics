## Checking prime characteristic

```python
N = 39916801
# N = 479001599
# N = 87178291199
```

```python
%%time
for n in range(2, N):
    if N % n == 0:
        print(f'{N} is not prime ({n}).')
        break
```

```python
%%time
n = 2
while n < N:
    if N % n == 0:
        print(f'{N} is not prime ({n}).')
        break
    n += 1
```
