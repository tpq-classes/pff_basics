## Mean-Variance Portfolios

### Financial Market

```python
B0, S0 = 10, 10
```

```python
B = np.array((11, 11))
```

```python
S = np.array((20, 5))
```

```python
M0 = np.array((B0, S0))
```

```python
M = np.array((B, S)).T
```

```python
M
```

```python
R = M / M0 - 1
```

```python
R
```

```python
p = 0.5
```

```python
P = np.array((p, 1 - p))
```

```python
np.dot(P, R)
```

### Portfolios

```python
s = 0.6
```

```python
b = 1 - s
```

```python
b
```

```python
np.dot((b, s), np.dot(P, R))
```

```python
def portfolio_return(phi):
    return np.dot(phi, np.dot(P, R))
```

```python
phi = (1 - s, s)
```

```python
portfolio_return(phi)
```

```python
R.std(axis=0)  # for equal probabilities
```

```python
s * R.std(axis=0)[1]
```

```python
def portfolio_volatility(phi):
    return phi[1] * R.std(axis=0)[1]
```

```python
portfolio_volatility(phi)
```

```python
s_ = np.linspace(0, 1, 21)
s_
```

```python
pr = [portfolio_return((1 - s, s)) for s in s_]
```

```python
pv = [portfolio_volatility((1 - s, s)) for s in s_]
```

```python
from pylab import plt
plt.style.use('seaborn-v0_8')
%config InlineBackend.figure_format = 'svg'
```

```python
plt.plot(s_, pr, label='portfolio return')
plt.plot(s_, pv, label='portfolio volatility')
plt.xlabel('s = stock position')
plt.legend();
```

```python
plt.plot(pv, pr)
plt.xlabel('portfolio volatility')
plt.ylabel('portfolio return');
```

## Two Risky Assets

### Financial Market

```python
T0, S0 = 10, 10
```

```python
T = np.array((7, 14))
```

```python
S = np.array((20, 5))
```

```python
M0 = np.array((T0, S0))
```

```python
M = np.array((T, S)).T
```

```python
M
```

```python
R = M / M0 - 1
```

```python
R
```

```python
p = 0.5
```

```python
P = np.array((p, 1 - p))
```

```python
np.dot(P, R)
```

### Portfolios

```python
s = 0.6
```

```python
b = 1 - s
```

```python
b
```

```python
np.dot((b, s), np.dot(P, R))
```

```python
def portfolio_return(phi):
    return np.dot(phi, np.dot(P, R))
```

```python
phi = (1 - s, s)
```

```python
portfolio_return(phi)
```

```python
R.std(axis=0)  # for equal probabilities
```

```python
np.cov(R, ddof=0)
```

```python
def portfolio_volatility(phi):
    return np.sqrt(np.dot(phi, np.dot(np.cov(R.T, ddof=0), phi)))
```

```python
phi = (0, 1)
```

```python
portfolio_volatility(phi)
```

```python
s_ = np.linspace(0, 1, 101)
s_[:10]
```

```python
pr = [portfolio_return((1 - s, s)) for s in s_]
```

```python
pv = [portfolio_volatility((1 - s, s)) for s in s_]
```

```python
plt.plot(s_, pr, label='portfolio return')
plt.plot(s_, pv, label='portfolio volatility')
plt.xlabel('s = stock position')
plt.legend();
```

```python
plt.plot(pv, pr)
plt.xlabel('portfolio volatility')
plt.ylabel('portfolio return');
```
