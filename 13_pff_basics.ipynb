## Visualization

```python
from pylab import plt
plt.style.use('seaborn-v0_8')
%config InlineBackend.figure_format = 'svg'
```

```python
plt.plot(rw.T);
```

```python
def gen_rw(M=50, I=1000):
    rn = rng.standard_normal((M + 1, I))
    rn[0] = 0.0
    rw = rn.cumsum(axis=0)
    return rw
```

```python
rw = gen_rw(I=100)
```

```python
plt.plot(rw);
```

```python
rw = gen_rw(I=100000)
```

```python
plt.hist(rw[-1], bins=35);
```

```python
rw[-1].mean()
```

```python
rw[-1].min()
```

```python
rw[-1].max()
```

```python
rw[-1].std()
```
