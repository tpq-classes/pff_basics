## Arbitrage Pricing

### Financial Market

```python
B0 = 10
B = np.array((11, 11))
```

```python
S0 = 10
S = np.array((20, 5))
```

```python
M0 = np.array((B0, S0))
M = np.array((B, S)).T
```

```python
M
```

### Replication 

```python
K = 14.5
```

```python
C = np.maximum(S - K, 0)
```

```python
# P = np.maximum(K - S, 0)
```

```python
C
```

```python
# P
```

```python
phi = np.linalg.solve(M, C)
```

```python
phi
```

```python
np.dot(M, phi)
```

```python
C0 = np.dot(phi, M0)
```

```python
C0
```

### Regression

```python
phi = np.linalg.lstsq(M, C, rcond=-1)[0]
```

```python
phi
```

```python
np.dot(phi, M0)
```

### Neural Net

```python
M = np.array((S, B)).T
```

```python
from numpy.random import default_rng
```

```python
rng = default_rng()
```

```python
# output layer
w = rng.random(2)
l1 = np.dot(M, w)
l1
```

```python
# deltas
d = l1 - C
d
```

```python
# mean-squared error
(d ** 2).mean()
```

```python
# learning rate
alpha = 0.025
```

```python
# update values
u = alpha * d
```

```python
# weights adjustments
w -= u
```

```python
l1 = np.dot(M, w)
l1
```

```python
alpha = 0.01
w = rng.random(2)
for _ in range(400):
    l1 = np.dot(M, w)
    d = l1 - C
    u = alpha * d
    w -= u
    if _ % 25 == 0:
        print((d ** 2).mean())
```

```python
l1
```

## Martingale Measure

```python
p = 0.5
```

```python
P = np.array((p, 1 - p))
```

```python
P
```

```python
P.sum()
```

```python
i = np.dot(B, P) / B0 - 1
i
```

```python
def E(P):
    return np.dot(S, P) / (1 + i)
```

```python
E(P)
```

```python
E(P) == S0
```

```python
from scipy.optimize import minimize
```

```python
bnds = ((0, 1), (0, 1))
```

```python
cons = {'type': 'eq', 'fun': lambda Q: Q.sum() - 1}
```

```python
def error(Q):
    return (E(Q) - S0) ** 2
```

```python
opt = minimize(error, P, bounds=bnds, constraints=cons)
```

```python
opt
```

```python
Q = opt['x'].round(6)
```

```python
Q
```

```python
E(Q)
```

```python
E(Q) == S0
```

```python
C
```

```python
C0 = np.dot(Q, C) / (1 + i)
```
