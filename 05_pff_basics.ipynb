```python
%%time
for n in range(2, N):
    if N % n == 0:
        print(f'{N} is not prime ({n}).')
        break
else:
    print(f'{N} is prime.')
```

```python
%%time
n = 2
while n < N:
    if N % n == 0:
        print(f'{N} is not prime ({n}).')
        break
    n += 1
```

## Algorithm 1

```python
def is_prime(N):
    for n in range(2, N):
        if N % n == 0:
            return False, n
    return True
```

```python
is_prime(133)
```

```python
133 / 7
```

```python
%time is_prime(N1)
```

## Algorithm 2

```python
def is_prime(N):
    if N % 2 == 0:
        return False, 2
    for n in range(3, N, 2):
        if N % n == 0:
            return False, n
    return True
```

```python
%time is_prime(N1)
```

```python
%time is_prime(N2)
```

## Algorithm 3

```python
def is_prime(N):
    if N % 2 == 0:
        return False, 2
    for n in range(3, int(N / 2), 2):
        if N % n == 0:
            return False, n
    return True
```

## Algorithm 4 

```python
import math
```

```python
def is_prime(N):
    if N % 2 == 0:
        return False, 2
    for n in range(3, int(math.sqrt(N)) + 1, 2):
        if N % n == 0:
            return False, n
    return True
```

## Algorithm 5

```python
def is_prime(N):
    if N % 2 == 0:
        return False, 2
    if N % 3 == 0:
        return False, 3
    for n in range(5, int(math.sqrt(N)) + 1, 6):
        if N % n == 0:
            return False, n
        if N % (n + 2) == 0:
            return False, n
    return True
```

```python
%time is_prime(N6)
```

```python
%timeit is_prime(N6)
```

## Algorithm 6

```python
def is_prime(N):
    if N % 2 == 0 or N % 3 == 0:
        return False
    for n in range(5, int(math.sqrt(N)) + 1, 6):
        if N % n == 0 or N % (n + 2) == 0:
            return False
    return True
```

```python
%time is_prime(N6)
```

```python
%timeit is_prime(N6)
```
