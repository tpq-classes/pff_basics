## Random Walks

```python
from numpy.random import default_rng
```

```python
rng = default_rng()
```

```python
rw = rng.standard_normal((101, 10))
```

```python
rw[0] = 0.0
```

```python
rw = rw.cumsum(axis=0)
```

```python
rw[:3]
```

```python
from pylab import plt
plt.style.use('seaborn-v0_8')
%config InlineBackend.figure_format = 'svg'
```

```python
plt.plot(rw);
```

```python
rw_df = pd.DataFrame(rw)
```

```python
rw_df.head()
```

```python
from string import ascii_lowercase
```

```python
rw.shape
```

```python
rw_df.columns = list(ascii_lowercase[:rw.shape[1]])
```

```python
rw_df.head()
```

```python
rw_df.index = pd.date_range('2023-1-1', freq='B', periods=rw.shape[0])
```

```python
rw_df.head()
```

```python
rw_df.plot();
```

```python
rw_df.plot(colormap='coolwarm');
```
