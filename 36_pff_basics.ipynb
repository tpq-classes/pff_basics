## Finance Environment

We need to implement the following:

* `class ActionSpace`
* `action_space.sample()`
* `class Finance`
* `finance.reset()`
* `finance.step()`

```python
import random
```

```python
class ActionSpace:
    def sample(self):
        return random.randint(0, 1)
```

```python
action_space = ActionSpace()
```

```python
for _ in range(10):
    print(action_space.sample(), end=' ')
```

```python
class Finance:
    url = 'https://certificate.tpq.io/mlfin.csv'
    def __init__(self, symbol, feature):
        self.symbol = symbol
        self.feature = feature
        self.action_space = ActionSpace()
        self.min_accuracy = 0.475
        self._get_data()
        self._prepare_data()
    def _get_data(self):
        self.raw = pd.read_csv(self.url, index_col=0, parse_dates=True)
    def _prepare_data(self):
        self.data = pd.DataFrame(self.raw[self.symbol]).dropna()
        self.data['r'] = np.log(self.data / self.data.shift(1))
        self.data['d'] = np.where(self.data['r'] > 0, 1, 0)
        self.data.dropna(inplace=True)
        self.data_ = (self.data - self.data.mean()) / self.data.std()
    def reset(self):
        self.bar = 4
        self.treward = 0
        state = self.data_[self.feature].iloc[self.bar - 4:self.bar].values
        return state, {}
    def step(self, action):
        correct = True if action == self.data['d'].iloc[self.bar] else False
        reward = 1 if correct else 0
        self.treward += reward
        self.bar += 1
        accuracy = self.treward / (self.bar - 4)
        if self.bar >= len(self.data):
            done = True
        elif reward == 1:
            done = False
        elif (accuracy < self.min_accuracy) and (self.bar > 15):
            done = True
        else:
            done = False
        next_state = self.data_[self.feature].iloc[self.bar - 4:self.bar].values
        return next_state, reward, done, False, {}
```
