## Approximation

### Functional Relationship

```python
x = np.linspace(-5, 5)
```

```python
y = x ** 2 / 2 + np.sin(x) * 2
```

```python
plt.plot(x, y);
```

```python
reg = np.polyfit(x, y, deg=3)
```

```python
p = np.polyval(reg, x)
```

```python
plt.plot(x, y)
plt.plot(x, p, 'ro');
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
from keras.layers import Dense
from keras.models import Sequential
from sklearn.metrics import mean_squared_error
```

```python
optimizer = keras.optimizers.Adam(learning_rate=0.001)
```

```python
model = Sequential()
model.add(Dense(128, activation='relu', input_dim=1))
model.add(Dense(1, activation='linear'))
# model.compile(loss='mse', optimizer='adam')
model.compile(loss='mse', optimizer=optimizer, metrics=['mse', 'mae'])
```

```python
%time res = model.fit(x, y, epochs=5000, verbose=False)
# warm start fitting (= starts from previous state)
```

```python
p = model.predict(x).flatten()
```

```python
plt.plot(x, y)
plt.plot(x, p, 'ro');
```

```python
model.evaluate(x, y)
```

```python
mean_squared_error(y, p)
```

```python
error = pd.DataFrame(res.history)
```

```python
error['mse'].iloc[2000:].plot();
```

```python
error['mae'].iloc[2000:].plot();
```

### Random Data

```python
x = np.linspace(-5, 5, 15)
```

```python
from numpy.random import default_rng
```

```python
rng = default_rng(100)
```

```python
y = rng.standard_normal(len(x)) * 2
```

```python
plt.plot(x, y);
```

```python
model = Sequential()
model.add(Dense(100, activation='relu', input_dim=1))
model.add(Dense(100, activation='relu'))
model.add(Dense(1, activation='linear'))
model.compile(loss='mse', optimizer=keras.optimizers.Adam(learning_rate=0.001),
             metrics=['mse', 'mae'])
```

```python
%time res = model.fit(x, y, epochs=10000, verbose=False)
```

```python
model.evaluate(x, y)
```

```python
p = model.predict(x).flatten()
```

```python
plt.plot(x, y)
plt.plot(x, p, 'ro');
```

```python
error = pd.DataFrame(res.history)
```

```python
error['mse'].iloc[2000:].plot();
```

```python
error['mae'].iloc[2000:].plot();
```
