```python
def rng(N=1000, dist='uniform'):
    if dist == 'uniform':
        rn = [random.random() for _ in range(N)]
    else:
        rn = [random.gauss(0, 1) for _ in range(N)]
    return rn
```

## Random Walk

```python
rn = rng(N=250, dist='gauss')
```

```python
rn[0] = 0.0
```

```python
plt.plot(rn)
plt.plot(rn, 'r.');
```

```python
rw = list()
rw.append(rn[0])
for i in range(1, len(rn)):
    rw.append(rw[i - 1] + rn[i])
```

```python
rw[:10]
```

```python
plt.plot(rw);
```

```python
def rwg(N=250):
    rn = rng(N=N, dist='gauss')
    rn[0] = 0.0
    rw = list()
    rw.append(rn[0])
    for i in range(1, len(rn)):
        rw.append(rw[i - 1] + rn[i])
    return rw
```

```python
rw = rwg()
```

```python
rw[0]
```

```python
plt.plot(rw)
plt.plot(rw, 'r.');
```

```python
rws = [rwg(N=100) for _ in range(100)]
```

```python
# rws
```

```python
for rw in rws:
    plt.plot(rw);
```
