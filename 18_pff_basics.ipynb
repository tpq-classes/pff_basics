## Random Walk

```python
N = 252
```

```python
rw = np.random.standard_normal(N)
```

```python
rw[0] = 0.0
```

```python
index = pd.date_range(start='2023-01-01', freq='B', periods=N)
```

```python
srw = pd.Series(rw.cumsum(), index=index)
```

```python
srw
```

```python
srw.head()
```

```python
srw.tail(3)
```
