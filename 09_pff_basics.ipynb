## Vectorized Operations

Vectorized operation = "no Python loop visible". In other words, the looping takes place behind the scenes, for example, based on highly optimized C code. The benefits generally are

* more concise code
* higher execution speed

```python
l = list(range(10))
l
```

```python
sum(l)
```

```python
def my_sum(x):
    s = 0
    for i in range(len(x)):
        s += x[i]
    return s
```

```python
my_sum(l)
```

```python
sq = [i ** 2 for i in l]
```

```python
sq
```

```python
2 * l  # two times the content of l
```

```python
a = np.array(l)
a
```

```python
2 * a  # vectorized operation = scalar product (element-wise multiplication)
```

```python
[2 * i for i in l]  # in Python a for loop is needed (= not vectorized)
```

```python
a / 2
```

## Speed & Memory

```python
import sys
```

```python
N = 10_000_000
```

```python
l = list(range(N))
```

```python
sys.getsizeof(l)
```

```python
a = np.arange(N)
```

```python
sys.getsizeof(a)
```

```python
%time sum(l)
```

```python
%time a.sum()
```

```python
import math
import random
```

```python
%timeit sq = [math.sqrt(i) for i in l]
```

```python
rnl = [random.random() for _ in range(N)]
```

```python
rnl[:6]
```

```python
sys.getsizeof(rnl)
```

```python
rna = np.array(rnl)
```

```python
sys.getsizeof(rna)
```

```python
%timeit ex = [math.exp(i) for i in rnl]
```

## Tradeoff Memory vs. Speed

```python
sys.getsizeof(range(N))
```

```python
%time sum(range(N))
```

```python
a = np.array(range(N))
```

```python
sys.getsizeof(a)
```

```python
%time a.sum()
```

```python
sys.getsizeof(a) / sys.getsizeof(range(N))
```
